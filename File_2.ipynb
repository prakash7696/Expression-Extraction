```python
import re

message = "I am interested in manufacturing PCBs 2 for my project. Attached are the gerber and design files. Kindly provide a quotation for approximately 100 unit. Also, please let me know the estimated manufacturing timeline with order 1234. We need fifty units."

expressions = ["units?", "PCBs?", "gerber", "order"]
textual_numbers = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten"]

# Construct the regular expression pattern for numerical numbers
pattern_numerical = r"\b(\d+)\s*(?:" + "|".join(expressions) + r")\b"

# Construct the regular expression pattern for textual numbers
pattern_textual = r"\b(" + "|".join(textual_numbers) + r")\s*(?:" + "|".join(expressions) + r")\b"

# Search for numerical number matches
matches_numerical = re.findall(pattern_numerical, message, re.IGNORECASE)
numbers_numerical = [int(match) for match in matches_numerical]

# Search for textual number matches
matches_textual = re.findall(pattern_textual, message, re.IGNORECASE)
numbers_textual = [textual_numbers.index(match.lower()) + 1 for match in matches_textual]

# Combine numerical and textual numbers
numbers = numbers_numerical + numbers_textual

print("Numbers:", numbers)
```

```python
import re
from word2number import w2n

message = "I am interested in manufacturing PCBs for my project. Attached are the gerber and design files. Kindly provide a quotation for approximately 100 unit. Also, please let me know the estimated manufacturing timeline with order 1234. We need fifty units."

expressions = ["units?", "PCBs?", "gerber"]

# Generate the list of all possible expressions for matching
all_expressions = expressions + [expression + "s?" for expression in expressions]

# Construct the regular expression pattern for numbers before expressions
pattern_before = r"\b((?:\d+|[a-zA-Z]+\s*[a-zA-Z]*))\s*(?:" + "|".join(all_expressions) + r")\b"

# Construct the regular expression pattern for numbers after expressions
pattern_after = r"(?:" + "|".join(all_expressions) + r")\s*((?:\d+|[a-zA-Z]+\s*[a-zA-Z]*))\b"

# Search for matches before expressions
matches_before = re.findall(pattern_before, message, re.IGNORECASE)
numbers_before = []
expressions_before = []
for match in matches_before:
    try:
        if match.isdigit():
            numbers_before.append(int(match))
            expressions_before.append(None)
        else:
            numbers_before.append(w2n.word_to_num(match))
            expression_match = re.search(r"\b(" + "|".join(all_expressions) + r")\b", match, re.IGNORECASE)
            if expression_match:
                expressions_before.append(expressions_match.group(1))
            else:
                expressions_before.append(None)
    except ValueError:
        pass

# Search for matches after expressions
matches_after = re.findall(pattern_after, message, re.IGNORECASE)
numbers_after = []
expressions_after = []
for match in matches_after:
    try:
        if match.isdigit():
            numbers_after.append(int(match))
            expressions_after.append(None)
        else:
            numbers_after.append(w2n.word_to_num(match))
            expression_match = re.search(r"\b(" + "|".join(all_expressions) + r")\b", match, re.IGNORECASE)
            if expression_match:
                expressions_after.append(expressions_match.group(1))
            else:
                expressions_after.append(None)
    except ValueError:
        pass

# Combine numbers and expressions before and after expressions
numbers = numbers_before + numbers_after
expressions = expressions_before + expressions_after

print("Numbers:", numbers)
print("Expressions:", expressions)
```

```python
import re
from word2number import w2n

message = "I am interested in manufacturing PCBs for my project. Attached are the gerber and design files. Kindly provide a quotation for approximately 100 unit. Also, please let me know the estimated manufacturing timeline with order 1234. We need fifty units."

expressions = ["units?", "PCBs?", "gerber"]

# Generate the list of all possible expressions for matching
all_expressions = expressions + [expression + "s?" for expression in expressions]

# Construct the regular expression pattern for numbers before expressions
pattern_before = r"\b((?:\d+|[a-zA-Z]+(?:\s+[a-zA-Z]+)*))\s*(?:" + "|".join(expressions) + r")\b"

# Construct the regular expression pattern for numbers after expressions
pattern_after = r"(?:" + "|".join(all_expressions) + r")\s*((?:\d+|[a-zA-Z]+(?:\s+[a-zA-Z]+)*))\b"

# Search for matches before expressions
matches_before = re.findall(pattern_before, message, re.IGNORECASE)
numbers_before = []
expressions_before = []
for match in matches_before:
    try:
        if match.isdigit():
            numbers_before.append(int(match))
        else:
            numbers_before.append(w2n.word_to_num(match))
            match_expressions = re.findall(r"\b(" + "|".join(expressions) + r")\b", match, re.IGNORECASE)
        if match_expressions:
            expressions_before.append(match_expressions[0])
    except ValueError:
        pass

# Search for matches after expressions
matches_after = re.findall(pattern_after, message, re.IGNORECASE)
numbers_after = []
expressions_after = []
for match in matches_after:
    try:
        if match.isdigit():
            numbers_after.append(int(match))
        else:
            numbers_after.append(w2n.word_to_num(match))
            match_expressions = re.findall(r"\b(" + "|".join(expressions) + r")\b", match, re.IGNORECASE)
        if match_expressions:
            expressions_after.append(match_expressions[0])
    except ValueError:
        pass

# Combine numbers and expressions before and after expressions
numbers = numbers_before + numbers_after
expressions = expressions_before + expressions_after

print("Numbers:", numbers)
print("Expressions:", expressions)
```

```python
import re
from word2number import w2n
import pandas as pd


with open("D:\DA\Innotrat LABS\Intent or Context Analysis\context_app.txt", 'r') as file:
    message = file.read().rstrip()

# message = """"""

expressions = ["unit", "PCB", "gerber", "quantity", "order was","order", "order no", "order no.", "order number"]

# Generate the list of all possible expressions for matching
all_expressions = expressions + [expression + "s?" for expression in expressions]

# Construct the regular expression pattern for numbers before expressions
pattern_before = r"\b((?:\d+|[a-zA-Z]+\s*[a-zA-Z]))\s((?:" + "|".join(all_expressions) + r"))\b"

# Construct the regular expression pattern for numbers after expressions
pattern_after = r"((?:" + "|".join(all_expressions) + r"))\s*((?:\d+|[a-zA-Z]+\s*[a-zA-Z]*))\b"

# Search for matches before expressions
matches_before = re.findall(pattern_before, message, re.IGNORECASE)
numbers_before = []
expressions_before = []
for match in matches_before:
    try:
        if match[0].isdigit():
            numbers_before.append(int(match[0]))
            expressions_before.append(match[1])
        else:
            numbers_before.append(w2n.word_to_num(match[0].lower()))
            expressions_before.append(match[1])
    except ValueError:
        pass

# Search for matches after expressions
matches_after = re.findall(pattern_after, message, re.IGNORECASE)
numbers_after = []
expressions_after = []
for match in matches_after:
    try:
        if match[1].isdigit():
            numbers_after.append(int(match[1]))
            expression = match[0]
            expressions_after.append(expression)
        else:
            numbers_after.append(w2n.word_to_num(match[1].lower()))
            expression = match[0]
            #if expression.endswith("s"):
                #expression = expression[:-1]  # Remove the 's' from the expression if present
            expressions_after.append(expression)
    except ValueError:
        pass

# Combine numbers and expressions before and after expressions
numbers = numbers_before + numbers_after
expressions = expressions_before + expressions_after

# print("Numbers:", numbers)
# print("Expressions:", expressions)

mapped_list = []

for item1, item2 in zip(expressions, numbers):
    item1_upper = item1.upper()
    mapped_list.append((item1_upper, item2))
    
for item1, item2 in mapped_list:
    print(f'{item1}: {item2}')
```
